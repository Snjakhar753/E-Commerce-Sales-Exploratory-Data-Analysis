## - Exploratory Data Analysis

**<font color= red> Order Priority</font>**

```python
# Analyise priority of orders of the product
df['Order Priority'].value_counts()
```

- <font color=green> Data cleaning </font>

```python
# find the unique values
df['Order Priority'].unique()
```

```python
df['Order Priority']= df['Order Priority'].replace('Critical ','Critical')
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='Order Priority',data=df,color="skyblue")
plt.title("Count of Order Priority")      
plt.show()
```

**<font color= red> Shipping Mode </font>**

```python
# # make a pie chart to show which category has high shiping
df['Ship Mode'].value_counts()
```

```python
x=df['Ship Mode'].value_counts().index
y=df['Ship Mode'].value_counts().values
```

```python
plt.figure(figsize=(5,4))
plt.pie(y,labels=x,startangle=90,autopct="%.2f%%")
plt.legend(loc=4)
plt.show()
```

- <font color = green>In which shiping mode which category of product and in how much amount</font>

```python
sns.countplot(x='Ship Mode', data= df,hue='Product Category')
plt.show()
```

**<font color= red> Order Date : Order per Year</font>**

**<font color=red> State wise order detail </font>**

```python
 # Top 5 state having highest sales 
top_5=df['State or Province'].value_counts().nlargest(5)
x = top_5.index
y = top_5.values
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x='State or Province',data=df,order=x,color="skyblue")
plt.title('Top 5 Highest sales state')
plt.show()
```

- <font color = green> highest order placed in California</font>
